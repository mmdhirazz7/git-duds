```python
# Importing libraries
import json
import time

# Loading the inventory records from JSON to string
fd = open("Records.json", "r")
js = fd.read()
fd.close()
#Converting from string to JSON/Dictionary
record = json.loads(js)

# Loading the sales records
fd = open("Sales.txt",'r')
sale = fd.readlines()
counter = len(sale)
fd.close()

# Displaying the MENU
print("---------------------------  M E N U  ----------------------------\n")

for key in record.keys():
    print(key ,":",record[key]["Name"],"|", record[key]["Price"],"|", record[key]["Qn"])

print("------------------------------------------------------------------\n")

# Taking user input about the product and user details
user_name = str(input("Enter your Name      :"))
user_mail = str(input("Enter mail ID        :"))
user_phn  = str(input("Enter your phone no. :"))
user_prid = str(input("Enter the Product ID :"))
user_qn   = int(input("Enter th Quantity    :"))

# If we have enough quantity for the user
if record[user_prid]["Qn"] >= user_qn:
    # Displaying purchase details
    print("-------------------------------------------------------------------")
    print("              Bill                                   \n")
    
    print("Name           : ", record[user_prid]["Name"])
    print("Price          : ", record[user_prid]["Price"],"Rs")
    print("Quantity       : ", user_qn)
    
    print("------------------------------------------------------------------\n")
    
    print("Billing Amount : ", user_qn * record[user_prid]["Price"],"Rs")

    # Billing amount
    bill = user_qn * record[user_prid]["Price"]
    # Discount on the purchase
    discount = 0
    if bill > 5000:
        discount = bill*0.1
        print("Discount       : ", discount, "Rs")
    else:
        print("Discount       : ", "0 Rs")

    # Displaying taxes and the net billing amount
    print("CGST           : ", bill * 0.09)
    print("SGST           : ", bill * 0.09)
    net_bill = bill + (bill * 0.09 * 2) - (discount)
    print("------------------------------------------------------------------\n")
    print("Net amount     : ",net_bill,"Rs")
    print("------------------------------------------------------------------\n")
    # Updating the record dictionary
    record[user_prid]["Qn"] = record[user_prid]["Qn"] - user_qn

    # Generating sales in csv format
    sales = str(counter+1)+","+user_name+","+user_mail+","+user_phn+","+user_prid+","+record[user_prid]["Name"]+","+str(user_qn)+","+str(record[user_prid]["Price"])+","+str(round(discount,2))+","+str(round((2*bill*0.09),2))+","+str(round((net_bill),2))+","+time.ctime()+"\n"

    fd = open("Sales.txt", "a")
    fd.write(sales)
    fd.close()

# If we don't have enough quantity for the user
else:

    # Displaying info about the quantity
    print("Sorry! We don't have enough quantity currently.")
    print("We only have ",record[user_prid]["Qn"]," only.")

    # If user wants to purchase the whole quantity.
    ch = input("Press Y/y to purchase.")

    
    if ch == "Y" or ch == 'y':
        
        user_qn = record[user_prid]["Qn"]

        # Displaying purchase details
        print("-------------------------------------------------------------------")
        print("              Bill                                   \n")
        
        print("Name           : ", record[user_prid]["Name"])
        print("Price          : ", record[user_prid]["Price"],"Rs")
        print("Quantity       : ", user_qn)
        
        print("------------------------------------------------------------------\n")
        
        print("Billing Amount : ", user_qn * record[user_prid]["Price"],"Rs")

        # Billing amount
        bill = user_qn * record[user_prid]["Price"]
        # Discount on the purchase
        discount = 0
        if bill > 5000:
            discount = bill*0.1
            print("Discount       : ", discount, "Rs")
        else:
            print("Discount       : ", "0 Rs")

        # Displaying taxes and the net billing amount
        print("CGST           : ", bill * 0.09)
        print("SGST           : ", bill * 0.09)
        net_bill = bill + (bill * 0.09 * 2) - (discount)
        print("------------------------------------------------------------------\n")
        print("Net amount     : ",net_bill,"Rs")
        print("------------------------------------------------------------------\n")

        # Udating record dictionary
        record[user_prid]["Qn"] = record[user_prid]["Qn"] - user_qn

        # Generating sales data in csv format
        sales = str(counter+1)+","+user_name+","+user_mail+","+user_phn+","+user_prid+","+record[user_prid]["Name"]+","+str(user_qn)+","+str(record[user_prid]["Price"])+","+str(round(discount,2))+","+str(round((2*bill*0.09),2))+","+str(round((net_bill),2))+","+time.ctime()+"\n"

        fd = open("Sales.txt", "a")
        fd.write(sales)
        fd.close()

    # If user pressed anything except Y or y
    else:
        print("Thanks.")


# Converting dictionary into a string
js = json.dumps(record)

# Saving records in JSON file
fd = open("Records.json",'w')
fd.write(js)
fd.close()

print()
print('--------------------------------------------')
print("      Thank you! Inventory updated.")
print('--------------------------------------------')
```
